# Import Packages

For now, we will assume no packages need to be installed in Python environment. To get started with this notebook, run the `processing.py` script to generate relative and absolute error values for the data uploaded to this repository on your local machine.

```python
import os
import matplotlib.pyplot as plt
import numpy as np
```

# Visualize Relative Errors By Trial Type

To choose a folder of outputs from `processing.py`, set the `folder` variable. To view a results from a specific movement type, set the `key` variable to one of the commented-out category types above.

```python
folder = 'Error_Files_Exp_Rel'

drop_jumps = ['DJ1', 'DJ2', 'DJ3', 'DJ4', 'DJ5']
drop_jumps_asym = ['DJAsym1', 'DJAsym2', 'DJAsym3', 'DJAsym4', 'DJAsym5']
squats = ['squats1']
squats_asym = ['squatsAsym1']
sts = ['STS1']
sts_weak = ['STSweakLegs1']
walking = ['walking1', 'walking2', 'walking3', 'walking4']
walking_ts = ['walkingTS1', 'walkingTS2', 'walkingTS3', 'walkingTS4']

trials = [drop_jumps, drop_jumps_asym, squats, squats_asym, sts, sts_weak, walking, walking_ts]
category_names = ['drop_jumps', 'drop_jumps_asym', 'squats', 'squats_asym', 'sts', 'sts_weak', 'walking', 'walking_ts']
```

```python
rel_errors_by_movement_exp = {}
for i, trial_category in enumerate(trials):
    cur_errors = []
    for trial in trial_category:
        for cam in range(5):
            if cam == 2:
                continue
            path = os.path.join(folder, str(trial)+'_cam'+str(cam)+'.npy')
            if not os.path.exists(path):
                continue
            cur_errors += list(np.load(path))
    if len(cur_errors):
        rel_errors_by_movement_exp[category_names[i]] = cur_errors
```

```python
# category_names = ['drop_jumps', 'drop_jumps_asym', 'squats', 'squats_asym', 'sts', 'sts_weak', 'walking', 'walking_ts']
key = 'drop_jumps'
```

```python
plt.hist(rel_errors_by_movement_exp[key])
plt.xlabel("Peak Rel Error (%)")
plt.ylabel("Frequency")
plt.title(key)
plt.show()
print(str("Av Peak Rel Error: "+str(np.round(np.mean(rel_errors_by_movement_exp[key]), decimals=2))+'%'))
```

```python
# session0 = ['drop_jumps', 'drop_jumps_asym', 'squats', 'squats_asym', 'sts', 'sts_weak']
session0 = ['walking', 'walking_ts']
session0_errors = []
for movement in session0:
    session0_errors += list(rel_errors_by_movement_exp[movement])

plt.hist(session0_errors)
plt.xlabel("Peak Rel Error (%)")
plt.ylabel("Frequency")
plt.title('Session 1')
plt.show()

session0_errors = np.array(session0_errors)
print(str("Mean overestimation: "+str(np.round(np.mean(session0_errors[session0_errors>=0]), decimals=2))+"%"))
print(str("Mean underestimation: "+str(np.round(np.mean(session0_errors[session0_errors<=0]), decimals=2))+"%"))
```

# Visualize Relative Errors by Camera Angle

To choose a folder of outputs from `processing.py`, set the `folder` variable. To choose a specific camera angle to view, set the `cam_angle` variable. Note that by default, camera angles 0,1,3,4 from the OpenCap video datasets are the only viewable angles.

```python
folder = 'Error_Files_Exp_Rel'
cam_angle = 3
rel_errors_by_camera_exp = {}
for cam in range(5):
    cur_errors = []
    for trial_category in trials:
        for trial in trial_category:
            if cam == 2:
                continue
            path = os.path.join(folder, str(trial)+'_cam'+str(cam)+'.npy')
            if not os.path.exists(path):
                continue
            cur_errors += list(np.load(path))
    if len(cur_errors):
        rel_errors_by_camera_exp[str(cam)] = cur_errors
```

```python
plt.hist(rel_errors_by_camera_exp[str(cam_angle)])
plt.xlabel("Peak Rel Error (%)")
plt.ylabel("Density")
plt.title("Camera Angle 3")
plt.show()
print(str("Av Peak Rel Error: "+str(np.round(np.mean(rel_errors_by_camera_exp[str(3)]), decimals=2))+'%'))
```

```python
# category_names = ['drop_jumps', 'drop_jumps_asym', 'squats', 'squats_asym', 'sts', 'sts_weak', 'walking', 'walking_ts']
ips_cam_rel_errors = list(rel_errors_by_camera_exp[str(0)]) + list(rel_errors_by_camera_exp[str(1)])
contra_cam_rel_errors = list(rel_errors_by_camera_exp[str(3)]) + list(rel_errors_by_camera_exp[str(4)])

plt.hist(ips_cam_rel_errors)
plt.xlabel("Peak Rel Error (%)")
plt.ylabel("Density")
plt.title("Ipsilateral Camera Angles")
plt.show()
print(str("Av Peak Rel Error: "+str(np.round(np.mean(ips_cam_rel_errors), decimals=2))+'%'))

plt.hist(contra_cam_rel_errors)
plt.xlabel("Peak Rel Error (%)")
plt.ylabel("Density")
plt.title("Contralateral Camera Angles")
plt.show()
print(str("Av Peak Rel Error: "+str(np.round(np.mean(contra_cam_rel_errors), decimals=2))+'%'))
```
